## Intro to numeric types
    These consists of float, in and complex numbers
    integers are immutable objects.
    when we add something to an integer, it won't be the same object that we are working with anymore.

    Floating point numbers:
    -  These are also immutable objects. it corresponds to double.
    - the methods are very similar to int. 

    complex numbers:
    - imaginary numbers with real parts and imaginary parts. for example, taking squareroot of negative numbers.
    - to do math on complex numbers you have to use the c math library. 
    - These are also immutable objects. we can't modify and have them created with the same object id.

## Intro to sequences and sets

    Python list vs python tuples

    - you can also create it through the function list.
    lists are mutable objects, changing the value and adding things to the list.
    tuples on the other hand are immutable objects.
        - (numbers) or tuple([numbers])
    
    python strings
    - strings are sequences and consists of characters. they are immutable. 
    functions don't modify inplace. you will have to capture it in another variable.
    There are alot of methods for you to modify the strings in place.
