# Command Line

Regex: 
Regular Expressions 
- they are way to specify search patterns for another string.
- They can be very complicated. 

Most have standard libraries for regular expressions.
## Basics:
- plain characters match exactly
- a dot `(.)` matches anything
- a caret `(^)` matches the beginning of a string
- a Dollar sign `($)` matches the end of a string
- a pipe `(|)` matches either expression on the left and the right side of the pipe.

we can also include spaces

## Brackets and Prantheses
- a bracket expression matches any character in the brackets
- a hyphen is used to make ranges
    - for example. [ab2] will match to any of a,b,2
    - [a-z] will match any lowercase letter.
    - [a-zA-Z] will match any letter no matter the sign.

A star `(*)` matches the previous character/bracket expression zero or more times
A plus `(+)` matches one or more times 
A question mark `(?)` matches zero or more times. Optional

## Character Classes
short hand for common things people want to search
`\b` word boundary. Anything at the begining or end of the world.
`\w` anything Alphanumeric
`\s` matches spaces and tabs
`\d` matches any digit.
to match any special characters you have to use a `\`

## Grep:

Find texts in files

grep [options] [pattern] [files(filespath)]

options:
- r (resursively)
- i ignore case of letters
- v print non-matching lines
- o print only parts of the line that match the pattern
- c count the number of matching lines in files.

## Find:
Find file based on name or other properties

find [starting point] [expression]

expression:
- type : type of object
- name: name matching
- iname: case insensitive name matching
- mtime: last modified time
- print: print the file path(default)
- exec: run a command for each file.
